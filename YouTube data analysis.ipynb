```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Global YouTube Statistics.csv',encoding='latin1')
df

grouped_by_country=df.groupby('Country')['subscribers'].sum().sort_values(ascending=False)
grouped_by_country

plt.figure(figsize=(15,6))

grouped_by_country.plot(kind='bar')
plt.title('total subcribers based on country')
plt.ylabel('subcribers')
plt.xlabel('country')
plt.show()
```

```python
top_10_categories=df['category'].value_counts().head(10)

plt.figure(figsize=(10,10))
plt.pie(top_10_categories,labels=top_10_categories.index,autopct='%1.1f%%',startangle=90)
plt.title('top 10 category')
plt.show()
```
